```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.pyplot import imread
import numpy as np
```

```python
img = imread("/content/m7med.jpeg")
img.shape
```

```python
img = img/255
img = img.mean(axis=2)
img
```

```python
plt.imshow(img,cmap="gray")
```

```python
from matplotlib import projections
def transform(p):
  p = p/100
  img_pca = PCA(n_components=p).fit(img)
  transform_img = img_pca.transform(img)
  projections = img_pca.inverse_transform(transform_img)
  plt.imshow(projections,cmap="gray")
```

```python
transform(90)
```
