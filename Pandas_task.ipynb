9) How many students have scored above 90 in all subjects?


```python
# Count the number of students who scored above 90 in all subjects
num_student_90=len(df[(df['math score']>90)&(df['reading score']>90)&(df['writing score']>90)])
print(num_student_90)
```

10)Create a new column 'pass_status' which contains 'Pass' if the student's total score is greater than or equal to 150, otherwise 'Fail'.


.apply(lambda x: 'Pass' if x >= 150 else 'Fail'):
This applies a function to each element in the 'total_score' column.
The function is defined using a **lambda function**:
lambda x: 'Pass' if x >= 150 else 'Fail':
This defines a lambda function that takes one argument x (which represents each value in the 'total_score' column).
If the value x is greater than or equal to 150 (x >= 150), it returns 'Pass'.
Otherwise, it returns 'Fail'.

```python
# Create a new column 'pass_status'
df['pass_status']=df['total_score'].apply(lambda x:'pass' if x>=150 else 'Fail')
```
